```python
import serial
import re
arduino_port = "COM5" #serial port of Arduino
baud = 9600 #arduino uno runs at 9600 baud
fileName="analog-data.csv" #name of the CSV file generated
ser = serial.Serial(arduino_port, baud)
print("Connected to Arduino port:" + arduino_port)
file = open(fileName, "a")
print("Created file")
#display the data to the terminal
samples = 1000 #how many samples to collect
print_labels = False
line = 0 #start at 0 because our header is 0 (not real data)
while line <= samples:
    # incoming = ser.read(9999)
    # if len(incoming) > 0:
    if print_labels:
        if line==0:
            print("Printing Column Headers")
        else:
            print(line)
    getData=str(ser.readline())
    data=getData[0:][:-2]
    print(data)

# file = open(fileName, "a")
    file.write(data) #write data with a newline
    line = line+1
print("Data collection complete!")
file.close()
```

```python
import re
import glob
for i in glob.glob("analog-data.csv"):
    reads = open(i, 'r')
    reader = reads.read()
    csvRe = re.sub("r", "" , str(reader))
    csvRe2 = re.sub("b'", "" , str(csvRe))
    csvRe3 = re.sub(",", ", " , str(csvRe2))
    csvRe4 = re.sub(r"\\", "" , str(csvRe3))
    writes = open("clean.csv", 'w')
    writes.write(csvRe4)
    reads.close()
    writes.close()
```

```python
import csv
import re
file = open('clean.csv')
read = csv.reader(file)
i=0
for j in read:
    if i <= 1000:
        csvRe5 = re.sub(" ", str(i) , str(read))
        writes = open("clean.csv", 'w')
        writes.write(csvRe5)
        i+=1      
file.close()
```
